## scale data based on overlapping lines

```python
## generate all pairs for 6 groups
from itertools import combinations
from scipy import stats
from scipy.stats import zscore
```

```python
df1 = pd.read_csv("Karthika_Mn_exp1_phenotype.csv",header=0)
df2 = pd.read_csv("Karthika_Mn_exp2_phenotype.csv",header=0)
df3 = pd.read_csv("Karthika_Mn_exp3_phenotype.csv",header=0)
df1.head()
```

```python
## add experiment column
df1['Group'] = "Exp1"
df2['Group'] = "Exp2"
df3['Group'] = "Exp3"
df1.columns.to_list()
```

```python
orderCols = ['Group',
    'NorthamID',
 'sowRL_mean_Ctl',
 'sowRL_mean_Mn',
 'harvestRL_mean_Ctl',
 'harvestRL_mean_Mn',
 'RTI_mean_Ctl',
 'RTI_mean_Mn',
 'ChangeRL_mean_Ctl',
 'ChangeRL_mean_Mn',
 'Stdry_perP_Ctl',
 'Stdry_perP_Mn',
 'Rtdry_perP_Ctl',
 'Rtdry_perP_Mn',
 'RRL',
 'RRTI',
 'RRtdry',
 'RStdry']
df1 = df1[orderCols]
df2 = df2[orderCols]
df3 = df3[orderCols]
```

```python
## join three experiments
# Concatenate DataFrames vertically (default axis=0)
df_merged = pd.concat([df1, df2, df3], ignore_index=True)
df_merged['NorthamID'] = df_merged['NorthamID'].astype(str) ## ID as strings
df_merged.head()
```

```python
# Check for trailing spaces using str.endswith
trailing_spaces_mask = df_merged['NorthamID'].str.startswith(' ')

# Count the number of rows with trailing spaces
trailing_spaces_mask.sum()
```

```python
group_list = ['Exp1','Exp2','Exp3']
res = list(combinations(group_list, 2))
res
```

```python
## check replicates in each group
group_dict = {}
for group in group_list:
    ID = df_merged[df_merged['Group'] == group]['NorthamID'].to_list()
    UID = list(set(ID))
    print(group, len(ID),len(UID)) ## group name, total lines, unique lines
    group_dict[group] = UID
```

```python
## count overlap between group pairs
def intersection(lst1, lst2):
    lst3 = [value for value in lst1 if value in lst2]
    return lst3

for pair in res:
    I = pair[0]
    J = pair[1]
    OVERLAP = intersection(group_dict[I],group_dict[J])
    print(I,J, len(OVERLAP))
```

```python
## all group compared to G1, leave out G3 which has no overlap with others
# Step1: groupby for each group to account for replicates in each group
# Step2: plot overlapping lines between groups, check for consistence in overlapping lines
# Step3: calculate ratio between groups
# Step4: normalize value based on average ratio
cols = ['Group',
 'ACCNAME',
 'sowRL_mean_Ctl',
 'sowRL_mean_Mn',
 'harvestRL_mean_Ctl',
 'harvestRL_mean_Mn',
 'RTI_mean_Ctl',
 'RTI_mean_Mn',
 'ChangeRL_mean_Ctl',
 'ChangeRL_mean_Mn',
 'Stdry_perP_Ctl',
 'Stdry_perP_Mn',
 'Rtdry_perP_Ctl',
 'Rtdry_perP_Mn',
 'RRL',
 'RRTI',
 'RRtdry',
 'RStdry']

df_G1 = df_merged[df_merged['Group'] == 'Exp1']
df_G2 = df_merged[df_merged['Group'] == 'Exp2']
df_G3 = df_merged[df_merged['Group'] == 'Exp3']


## compare to G1
df_G1G2 = df_G1.merge(df_G2,on='ACCNAME',how='inner')
df_G1G3 = df_G1.merge(df_G3,on='ACCNAME',how='inner')
```

## across-group scaling adjustment
## zscore for each group
## merge

```python
df_G1.head()
```

```python
## calculate ratio 
def cal_ratio(DF1,Trait):
    Y = Trait + "_y"
    X = Trait + "_x"
    ratios = DF1[Y] / DF1[X]
    return ratios.mean(), ratios.std()

# adjust G2, G3 to G1
Traits = ['sowRL_mean_Ctl',
 'sowRL_mean_Mn',
 'harvestRL_mean_Ctl',
 'harvestRL_mean_Mn',
 'RTI_mean_Ctl',
 'RTI_mean_Mn',
 'ChangeRL_mean_Ctl',
 'ChangeRL_mean_Mn',
 'Stdry_perP_Ctl',
 'Stdry_perP_Mn',
 'Rtdry_perP_Ctl',
 'Rtdry_perP_Mn',
 'RRL',
 'RRTI',
 'RRtdry',
 'RStdry']

for TRAIT in Traits:
    ## scale by average ratio
    df_G2[TRAIT] = df_G2[TRAIT] / cal_ratio(df_G1G2, TRAIT)[0]
    df_G3[TRAIT] = df_G3[TRAIT] / cal_ratio(df_G1G3, TRAIT)[0]

## merge adjusted value
## df_merged = pd.concat([df1, df2, df3], ignore_index=True)
df_adjust_merged = pd.concat([df_G1,df_G2,df_G3]).dropna().reset_index(drop=True)
df_adjust_merged.head()
```

```python
## zscore normalization
for TRAIT in Traits:
    df_adjust_merged[TRAIT] = stats.zscore(df_adjust_merged[TRAIT])
```

```python
df_adjust_merged.head()
```

```python
## calculate mean for overlapping lines
df_final = df_adjust_merged.groupby('ACCNAME').mean().reset_index()
df_final.head()
```
