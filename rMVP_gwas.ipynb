```r
# Full-featured function (Recommended)
MVP.Data(fileVCF="Karthika_Mn_threeExp.genotype",
         #filePhe="Phenotype.txt",
         fileKin=FALSE,
         filePC=FALSE,
         out="rmvp"
         )
```

```r
genotype <- attach.big.matrix("rmvp.geno.desc")
map <- read.table("rmvp.geno.map" , head = TRUE)
phenotype <- read.table("Karthika_Mn_threeExp.phenotype",head=TRUE)
```

```r
## calculate kinship and principle components from genotype data
MVP.Data.Kin(TRUE, mvp_prefix='rmvp', out='rmvp')
Kinship <- attach.big.matrix("rmvp.kin.desc")

MVP.Data.PC(TRUE, mvp_prefix='rmvp', out='rmvp', pcs.keep=5)
Covariates_PC <- bigmemory::as.matrix(attach.big.matrix("rmvp.pc.desc"))
```

```r
## run GWAS
for(i in 2:ncol(phenotype)){
  imMVP <- MVP(
    phe=phenotype[, c(1, i)],
    geno=genotype,
    map=map,
    K=Kinship,
    CV.GLM=Covariates_PC,
    CV.MLM=Covariates_PC,
    CV.FarmCPU=Covariates_PC,
    nPC.GLM=5,
    nPC.MLM=5,
    nPC.FarmCPU=5,
    priority="speed",
    #ncpus=10,
    vc.method="BRENT",
    maxLoop=10,
    method.bin="static",
    #permutation.threshold=TRUE,
    #permutation.rep=100,
    threshold=c(0.05),
    method=c("GLM", "MLM", "FarmCPU"),
    #file.output=c("pmap", "pmap.signal", "plot")
    file.output=TRUE
  )
  gc()
}
```
